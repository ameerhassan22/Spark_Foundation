## EDA

```python
# Load Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression  as LR
from sklearn.model_selection import train_test_split as split
```

```python
# Load Dataset
train = pd.read_csv('student_scores.csv')
train.head(10)
```

```python
# Show Missing Values
print(train.isnull().sum())
```

```python
plt.rcParams['figure.figsize']=(10,8)
sns.scatterplot(data=train,x='Hours',y='Scores')
plt.title('Scores for Hours',color='Maroon',fontsize=25)
```

```python
# Split Data
x = train.drop(columns='Scores',axis=0)
y= train['Scores']
x_train,x_test,y_train,y_test = split(x,y,random_state=42,test_size=0.2)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape,sep='\n')
# Linear Regression
model = LR()
```

```python
# Train Model
model.fit(x_train,y_train)
# Accuracy of Train and Test Data
AC1 = model.score(x_train,y_train)
AC2 = model.score(x_test,y_test)
print(f'Train Accuracy : {AC1*100:.2f} %')
print(f'Test Accuracy : {AC2*100:.2f} %')
```

```python
# Predict Score if student studied 9.25 hours
data = np.array([[9.25]])
print(model.predict(data))
```
